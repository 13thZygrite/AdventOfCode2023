```python
import itertools
import functools

def check(pattern, checksum):
    grouped = itertools.groupby(pattern.lstrip("."))
    lengths = [len(list(g)) for _, g in grouped]
    even_lengths = lengths[::2]
    if checksum == even_lengths:
        # print(checksum, even_lengths, odd_lengths)
        return True
    else:
        return False
    
# print(check("####...#..#.", [3, 2, 1]))
    
def check_initial(pattern, checksum):
    if ((pattern.count("?") + pattern.count("#")) < sum(checksum)):
        return False
    init_seg = pattern.split("?")[0]
    return recur_check(init_seg, tuple(checksum))

def recur_check(seg, checksum):
    if len(seg) == 0:
        return True
    if len(seg) > 0 and len(checksum) == 0:
        return False

    count = 0
    for char in seg:
        if char != "#":
            break
        else:
            count += 1
    if (count == len(seg)):
        if count <= checksum[0]:
            return True
        else:
            return False
    elif count != checksum[0]:
        return False
    else:
        return recur_check(seg.lstrip("#").lstrip("."), checksum[1::])
    
print(check_initial("???.###", [1, 1, 3])) # TRUE

# print(check_initial("#??.###", [1, 1, 3])) # True
# print(check_initial(".??.###", [1, 1, 3])) # True

# print(check_initial("##?.###", [1, 1, 3])) # false
# print(check_initial("#.?.###", [1, 1, 3])) # true
# print(check_initial(".#?.###", [1, 1, 3])) # true
# print(check_initial("..?.###", [1, 1, 3])) # true

# print(check_initial("###.###", [1, 1, 3])) # false
# print(check_initial("##..###", [1, 1, 3])) # false
# print(check_initial("#.#.###", [1, 1, 3])) # true
# print(check_initial("#...###", [1, 1, 3])) # false
# print(check_initial(".##.###", [1, 1, 3])) # false
# print(check_initial(".#..###", [1, 1, 3])) # false
# print(check_initial("..#.###", [1, 1, 3])) # false
# print(check_initial("....###", [1, 1, 3])) # false





    
```

```python
# Pure brute force

sum1 = 0

for line in input.splitlines():
    pattern = line.split()[0]
    checksum = [int(c) for c in line.split()[1].split(",")]
    length = pattern.count("?")
    for combo in itertools.product(".#", repeat=length):
        new_pattern = ""
        count = 0
        for char in pattern:
            if char == "?":
                new_pattern += combo[count]
                count += 1
            else:
                new_pattern += char
        if check(new_pattern, checksum):
            sum1 += 1

print(sum1)
```
