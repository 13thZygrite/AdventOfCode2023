```python
from functools import cache

@cache
def collision(brick1, brick2):
    assert(brick1[0][0] == brick1[1][0] or brick1[0][1] == brick1[1][1])
    assert(brick2[0][0] == brick2[1][0] or brick2[0][1] == brick2[1][1])
    #dumbest way possible
    brick1_points = set()
    brick2_points = set()
    for x in range(min(brick1[0][0],brick1[1][0]), max(brick1[0][0],brick1[1][0]) + 1):
        for y in range(min(brick1[0][1],brick1[1][1]), max(brick1[0][1],brick1[1][1]) + 1):
            brick1_points.add((x,y))
    for x in range(min(brick2[0][0],brick2[1][0]), max(brick2[0][0],brick2[1][0]) + 1):
        for y in range(min(brick2[0][1],brick2[1][1]), max(brick2[0][1],brick2[1][1]) + 1):
            brick2_points.add((x,y))
    return len(brick1_points.intersection(brick2_points)) > 0
```

```python
bricks = []

# This might be too slow but whatever

for line in input.splitlines():
    line.split("~")
    x1,y1,z1 = line.split("~")[0].split(",")
    x2,y2,z2 = line.split("~")[1].split(",")
    x1 = int(x1)
    y1 = int(y1)
    z1 = int(z1)
    x2 = int(x2)
    y2 = int(y2)
    z2 = int(z2)
    brick = ((x1, y1, z1), (x2, y2, z2)) if z1 <= z2 else ((x2, y2, z2), (x1, y1, z1))
    bricks.append(brick)

bricks.sort(key=lambda b: b[0][2])

fallen_bricks = []
for brick in bricks:
    highest_z = 0
    for lower_brick in fallen_bricks:
        if lower_brick[0][2] >= brick[0][2]:
            continue
        else:
            if collision(lower_brick, brick):
                #assert(lower_brick[1][2] >= highest_z)
                highest_z = max(highest_z,lower_brick[1][2])
            else:
                continue
    new_brick = ((brick[0][0], brick[0][1], highest_z + 1), (brick[1][0], brick[1][1], highest_z + 1 + (brick[1][2] - brick[0][2])))
    fallen_bricks.append(new_brick)
    fallen_bricks.sort(key=lambda b: b[0][2])

# print(bricks)
# print(fallen_bricks)

supports = {}
total = 0

for brick in fallen_bricks:
    support = []
    for brick2 in fallen_bricks:
        if brick2[0][2] == brick[1][2] + 1:
            if collision(brick, brick2):
                support.append(brick2)
    supports[brick] = support
# print(supports)

for brick in fallen_bricks:
    safe = True
    for supp in supports[brick]:
        count = 0
        for v in supports.values():
            if supp in v:
                count += 1
        if count == 1:
            safe = False
            break
    if safe:
        total += 1

print("Part 1: ", total)
print(len(bricks))

```
