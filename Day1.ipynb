```python
def find_digit_forwards(line):
    digit = ""
    for index,char in enumerate(line):
        if char in "0123456789":
            digit = char
            break
        if index >= 2:
            if line[index - 2: index +1] == "one":
                digit = "1"
                break
            if line[index - 2: index +1] == "two":
                digit = "2"
                break
            if line[index - 2: index +1] == "six":
                digit = "6"
                break
        if index >= 3:
            if line[index - 3: index +1] == "four":
                digit = "4"
                break
            if line[index - 3: index +1] == "five":
                digit = "5"
                break
            if line[index - 3: index +1] == "nine":
                digit = "9"
                break
        if index >= 4:
            if line[index - 4: index +1] == "three":
                digit = "3"
                break
            if line[index - 4: index +1] == "seven":
                digit = "7"
                break
            if line[index - 4: index +1] == "eight":
                
                digit = "8"
                break
    assert(digit != "")
    return digit

def find_digit_backwards(line):
    digit = ""
    reversed = line[::-1]
    for index,char in enumerate(reversed):
        if char in "0123456789":
            digit = char
            break
        if index >= 2:
            if reversed[index - 2: index +1] == "eno":
                digit = "1"
                break
            if reversed[index - 2: index +1] == "owt":
                digit = "2"
                break
            if reversed[index - 2: index +1] == "xis":
                digit = "6"
                break
        if index >= 3:
            if reversed[index - 3: index +1] == "ruof":
                digit = "4"
                break
            if reversed[index - 3: index +1] == "evif":
                digit = "5"
                break
            if reversed[index - 3: index +1] == "enin":
                digit = "9"
                break
        if index >= 4:
            if reversed[index - 4: index +1] == "eerht":
                digit = "3"
                break
            if reversed[index - 4: index +1] == "neves":
                digit = "7"
                break
            if reversed[index - 4: index +1] == "thgie":
                digit = "8"
                break
    assert(digit != "")
    return digit
```
