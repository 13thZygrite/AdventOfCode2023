```python
def right_90(dir):
    if dir == (1, 0):
        return (0,1)
    if dir == (0, 1):
        return (-1,0)
    if dir == (-1, 0):
        return (0,-1)
    if dir == (0, -1):
        return (1,0)

def left_90(dir):
    if dir == (1, 0):
        return (0,-1)
    if dir == (0, -1):
        return (-1,0)
    if dir == (-1, 0):
        return (0,1)
    if dir == (0, 1):
        return (1,0)
```

```python
from queue import PriorityQueue
from collections import defaultdict
array = {}

max_i = -1
max_j = -1

for j, line in enumerate(input.splitlines()):
    max_j = max(max_j, j)
    for i, char in enumerate(line):
        max_i = max(max_i, i)
        array[(i,j)] = int(char)

# I guess we A* now bois

start = (0,0)
goal = (max_i, max_j)

# State = score,current_node, curr_dir, count
# heuristic can't overestimate maybe just manhattan?

# Assumption that the path never loops, might questionable
nodes_to_find = PriorityQueue()
nodes_to_find.put((start, (1,0), 0), 0)
nodes_to_find.put((start, (0,1), 0), 0)

best_previous = {}

gmap = defaultdict(lambda: 100000000000000)
gmap[(start, (1,0), 0)] = 0

fmap = defaultdict(lambda: 100000000000000)
fmap[(start, (1,0), 0)] = max_i + max_j

end_goals = []


while len(nodes_to_find.queue) > 0:
    current = nodes_to_find.get()
    if current[0] == goal:
        end_goals.append(gmap[current])
    
    to_check = []
    if (current[0][0] + right_90(current[1])[0], current[0][1] + right_90(current[1])[1]) in array:
        to_check.append(((current[0][0] + right_90(current[1])[0], current[0][1] + right_90(current[1])[1]), right_90(current[1]), 1))
    if (current[0][0] + left_90(current[1])[0], current[0][1] + left_90(current[1])[1]) in array:
        to_check.append(((current[0][0] + left_90(current[1])[0], current[0][1] + left_90(current[1])[1]), left_90(current[1]), 1))
    if (current[0][0] + current[1][0], current[0][1] + current[1][1]) in array and current[2] <= 2:
        to_check.append(((current[0][0] + current[1][0], current[0][1] + current[1][1]), current[1], current[2] + 1))
    
    for check in to_check:
        tentative_g = gmap[current] + array[check[0]]
        if tentative_g < gmap[check]:
            best_previous[check] = current
            gmap[check] = tentative_g
            fmap[check] = tentative_g + (max_i - check[0][0] + max_j - check[0][1])
            if check not in nodes_to_find.queue:
                nodes_to_find.put(check, fmap[check])


print("Part 1: ", min(end_goals))


```
